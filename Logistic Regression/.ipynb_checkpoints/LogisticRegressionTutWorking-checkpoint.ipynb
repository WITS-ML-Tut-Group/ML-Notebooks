```python
# Define 2 2D Guassian distributions

# Class 0
x = [[], []]

x[0] = np.random.normal(-1, 1, 20)
x[1] = np.random.normal(1, 1, 20)
class0_points = np.array(x)

#Class 1
y = [[], []]

y[0] = np.random.normal(1, 1, 20)
y[1] = np.random.normal(-1, 1, 20)
class1_points = np.array(y)


print("Class 0")
print(class0_points)
print("Class 1")
print(class1_points)
```

```python
#Question 2(a)

theta = np.random.uniform(-0.5,0.5,3) #Initialising random theta values
print(theta)
    
x = np.linspace(-6,6,100)
y = -theta[1]/theta[2]*x - theta[0]/theta[2] #Initialising first iteration decision boundary
plt.plot(x, y)

plt.scatter(class0_points[0], class0_points[1])

plt.scatter(class1_points[0], class1_points[1])

plt.axis('equal')
plt.show()
```

```python
#Question 2(b)

def sigmoid(x): #Sigmoid function
    return 1/(1+np.exp(-x))

def prediction(theta, x): #Intermediate function
    tran = np.transpose(theta)
    return(sigmoid(np.dot(tran,x)))

sum = 0
correctTotal0 = 0
correctTotal1 = 0

for i in range(20): #Calculating log likelehood for class 0
    x = [1,class0_points[0][i],class0_points[1][i]]
    sum = sum + np.log(1-prediction(theta, x))
    predictedValue = round(prediction(theta, x),0)
    if(predictedValue == 0):
        correctTotal0 += 1
    
for i in range(20): #Calculating log likelehood for class 1
    x = [1,class1_points[0][i],class1_points[1][i]]
    sum = sum + np.log(prediction(theta, x))
    predictedValue = round(prediction(theta, x),0)
    if(predictedValue == 1):
        correctTotal1 += 1
        
#Printing out confusion matrix
    
print("Classes :  0      1")
print("     0    ",correctTotal0,"   ",20-correctTotal0)
print("     1    ",correctTotal1,"   ",20-correctTotal1)
print("")

sum = sum*-1
print(sum) # Printing out log likelihood
```

```python
# Question 2E

#train for class 0
print(theta)
for i in range(20):
    x = [1,class0_points[0][i],class0_points[1][i]]
    theta[0] = theta[0] - 0.01*(0 - prediction(theta,x))
    theta[1] = theta[1] - 0.01*(0 - prediction(theta,x))*x[1]
    theta[2] = theta[2] - 0.01*(0- prediction(theta,x))*x[2]
print(theta)
##train for class 1
for i in range(20):
    x = [1,class1_points[0][i],class1_points[1][i]]
    theta[0] = theta[0] - 0.01*(1 - prediction(theta,x))
    theta[1] = theta[1] - 0.01*(1 - prediction(theta,x))*x[1]
    theta[2] = theta[2] - 0.01*(1- prediction(theta,x))*x[2]
print(theta)
```

```python
x = np.linspace(-6,6,100)
y = -theta[1]/theta[2]*x - theta[0]/theta[2] #Initialising first iteration decision boundary
print(theta)
plt.plot(x, y)

plt.scatter(class0_points[0], class0_points[1])

plt.scatter(class1_points[0], class1_points[1])

plt.axis('equal')
plt.show()
```

```python
## error of current model
##training error
def likehood(x,y,theta,prediction):
    return y *np.log(prediction(theta,x)) + (1-y)*np.log(1- prediction(theta,x))


sum = 0
correctTotal0 = 0
correctTotal1 = 0

for i in range(20): #Calculating log likelehood for class 0
    x = [1,class0_points[0][i],class0_points[1][i]]
    sum = sum + likehood(x,0,theta,prediction)
    print(prediction(theta,x))
    predictedValue = round(prediction(theta,x),0)
    print(predictedValue)
    if(predictedValue == 0):
        correctTotal0 += 1
    
    x = [1,class1_points[0][i],class1_points[1][i]]
    sum = sum + likehood(x,1,theta,prediction)
    print(prediction(theta,x))
    predictedValue = round(prediction(theta,x),0)
    print(predictedValue)
    if(predictedValue == 1):
        correctTotal1 += 1
        
#Printing out confusion matrix
    
print("Classes :  0      1")
print("     0    ",correctTotal0,"   ",20-correctTotal0)
print("     1    ",20-correctTotal1,"   ",correctTotal1)
print("")

sum = sum*-1
print(sum) # Printing out log likelihood
```

```python
def diff(new,old):
    finish = [0,0,0]
    for i in range(3):
        finish[i] = abs(new[i] - old[i])
    sum= finish[0] + finish[1] + finish[2]
    return sum 

def myprediction(theta,x):
    return(sigmoid(np.dot(theta,x)))
    

condition = True
count = 0
newtheta = [0,0,0]
while condition == True and count < 1000:
    
    for i in range(20):
        
        newtheta = [theta[0],theta[1],theta[2]]
        x = [1,class0_points[0][i],class0_points[1][i]]
        newtheta[0] = newtheta[0] + 0 - myprediction(newtheta,x)
        newtheta[1] = newtheta[1] + 0.01*(0 - myprediction(newtheta,x))*x[1]
        newtheta[2] = newtheta[2] + 0.01*(0- myprediction(newtheta,x))*x[2]
        count = count +1
        con = diff(newtheta,theta)
        theta = [newtheta[0],newtheta[1],newtheta[2]]
        print("this is con ")
        print(con)
        if(con < 0.05):
            condition = False
            break
        
    ##train for class 1

        newtheta = [theta[0],theta[1],theta[2]]
        x = [1,class1_points[0][i],class1_points[1][i]]
        newtheta[0] = newtheta[0] + 1 - myprediction(newtheta,x)
        newtheta[1] = newtheta[1] + 0.01*(1 - myprediction(newtheta,x))*x[1]
        newtheta[2] = newtheta[2] + 0.01*(1- myprediction(newtheta,x))*x[2]
        count = count +1
        con = diff(newtheta,theta)
        theta = [newtheta[0],newtheta[1],newtheta[2]] 
        print("this is conn")
        print(con)
        if(con < 0.05):
            condition = False
            break
          

```
